```python
# Step 3: Data Reduction

# Some columns or variables can be dropped if they do not add value to our analysis.
# In our dataset, the column YRQTR has not relative data at the end of the data set. 
# Remove all rows from index 44 to the end (index 47).
CreditCardData = CreditCardData.drop(CreditCardData.index[44:])
CreditCardData.tail()
```

```python
# Remove all columns from  'Account Balance (Active Accounts Only) (50th percentile)' 
# to '90+ Days Past Due Rates: Balances Based'.
# ix or loc removing columns between two column names. 
#CreditCardData.drop(CreditCardData.ix[:, 'B':'D'].columns, axis=1)
# or Remove all columns between column index 
#CreditCardData.drop(CreditCardData.iloc[:, 1:3], inplace=True, axis=1)
 
CreditCardData = CreditCardData.drop(CreditCardData.loc[:, 'Account Balance (Active Accounts Only) (50th percentile)':'90+ Days Past Due Rates: Balances Based'].columns, axis=1)
CreditCardData.info()
```

```python
# # We start our Feature Engineering as we need to add some columns required for analysis.

# Step 4: Feature Engineering

# Feature engineering refers to the process of using domain knowledge to select 
# and transform the most relevant variables from raw data when creating a predictive model 
# using machine learning or statistical modeling. 
# The main goal of Feature engineering is to create meaningful data from raw data.
```

```python
CreditCardData.head()
```

```python
# Rename columns

CreditCardData.rename(columns={'Share of Accounts Making the Minimum Payment': 'Share of Accounts Making the Minimum Payment (%Percentage)'},inplace=True)
CreditCardData.rename(columns={'Share of Accounts Making Greater Than the Minimum Payment but Less Than the Full Balance': 'Share of Accounts Making Greater Than the Minimum Payment but Less Than the Full Balance (%Percentage)'},inplace=True)
CreditCardData.rename(columns={'Share of Accounts Making Full Balance Payment': 'Share of Accounts Making Full Balance Payment (%Percentage)'},inplace=True)

CreditCardData.head()
```

```python
# Step 6: Data Cleaning/Wrangling
# Remove $, %, , symbols in the dataframe and change data type from 'object' to 'float'

CreditCardData['Total Balances ($Billions)'] = CreditCardData['Total Balances ($Billions)'].str.replace('$', '').astype(float)
CreditCardData['Total Commitments ($Billions)'] = CreditCardData['Total Commitments ($Billions)'].replace('[\$,]', '', regex=True).astype(float)
CreditCardData['Share of Accounts Making the Minimum Payment (%Percentage)'] = CreditCardData['Share of Accounts Making the Minimum Payment (%Percentage)'].str.replace('%', '').astype(float)
CreditCardData['Share of Accounts Making Greater Than the Minimum Payment but Less Than the Full Balance (%Percentage)'] = CreditCardData['Share of Accounts Making Greater Than the Minimum Payment but Less Than the Full Balance (%Percentage)'].str.replace('%', '').astype(float)
CreditCardData['Share of Accounts Making Full Balance Payment (%Percentage)'] = CreditCardData['Share of Accounts Making Full Balance Payment (%Percentage)'].str.replace('%', '').astype(float)
CreditCardData['Revolving Balances Only ($Billions)'] = CreditCardData['Revolving Balances Only ($Billions)'].str.replace('$', '').astype(float)


# Display the updated DataFrame
print(CreditCardData.head())
```

```python

# Step 5: Creating Features

# To see what type of credit balance debt people have, we need to have two separate columns
# 'Revolving credit' and 'Installment credit'.
# We have 'Revolving Balances Only ($Billions)' column in our data set and 'Total Balances ($Billions)'.

# Revolving credit is when you can borrow money up to a certain limit and pay it back as you wish, like using a credit card.
# Installment credit is a loan with a set amount of money that you pay back in fixed payments over time, like a car loan or a mortgage.



# Add the column 'Installment Balances Only ($Billions)'
# using formula 'Total Balances ($Billions)' - 'Revolving Balances Only ($Billions)'.

CreditCardData['Installment Balances Only ($Billions)']=CreditCardData['Total Balances ($Billions)']-CreditCardData['Revolving Balances Only ($Billions)']
CreditCardData.head()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

CreditCardData['YRQTR'] = pd.to_datetime(CreditCardData['YRQTR'], format='%YQ%q')  # Convert 'YRQTR' to datetime format for better plotting

# Filter DataFrame for only the fourth quarter of each year
df_q4 = CreditCardData[CreditCardData['YRQTR'].dt.quarter == 4]

# Sort DataFrame by 'YRQTR' for better plotting
df_q4 = df_q4.sort_values('YRQTR')

# Calculate percentage change for the fourth quarter of each year
#The pct_change() method returns a DataFrame with the percentage difference between the values for each row and, by default, the previous row.
df_q4['Percentage Change'] = df_q4['Total Balances ($Billions)'].pct_change() * 100

# Filled Area Plotting
plt.figure(figsize=(10, 6))

# Highlight positive changes in deepskyblue and negative changes in lightcoral
plt.fill_between(df_q4['YRQTR'], df_q4['Percentage Change'], where=(df_q4['Percentage Change'] >= 0), color='deepskyblue', interpolate=True, label='Positive Change')
plt.fill_between(df_q4['YRQTR'], df_q4['Percentage Change'], where=(df_q4['Percentage Change'] < 0), color='lightcoral', interpolate=True, label='Negative Change')

# Adding labels and title
plt.xlabel('Year and Quarter')
plt.ylabel('Percentage Change')
plt.title('Percentage Change in Total Balances (Q4 Only) Over Time')

# Add legend
plt.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

CreditCardData['YRQTR'] = pd.to_datetime(CreditCardData['YRQTR'], format='%YQ%q')  # Convert 'YRQTR' to datetime format for better plotting

# Extract the year from the datetime column
CreditCardData['Year'] = CreditCardData['YRQTR'].dt.year

# Grouped Bar Plotting by Year
plt.figure(figsize=(12, 6))

bar_width = 0.25  # Adjust the width based on your preference
bar_positions = np.arange(len(CreditCardData['Year'].unique()))

plt.bar(bar_positions - bar_width, CreditCardData.groupby('Year')['Share of Accounts Making Full Balance Payment (%Percentage)'].mean(), 
        width=bar_width, color='green', label='Full Balance Payment')
plt.bar(bar_positions, CreditCardData.groupby('Year')['Share of Accounts Making Greater Than the Minimum Payment but Less Than the Full Balance (%Percentage)'].mean(), 
        width=bar_width, color='orange', label='Greater Than Min Payment')
plt.bar(bar_positions + bar_width, CreditCardData.groupby('Year')['Share of Accounts Making the Minimum Payment (%Percentage)'].mean(), 
        width=bar_width, color='blue', label='Min Payment')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Average Percentage (%)')  # Using average percentage for each year
plt.title('Average Share of Accounts Payment Distribution Over Years')

# Adjust x-axis ticks and labels
plt.xticks(bar_positions, CreditCardData['Year'].unique())

plt.legend()
plt.show()
```
